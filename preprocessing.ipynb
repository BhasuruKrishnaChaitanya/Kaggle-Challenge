```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df.head()
df['type'] = df['type'].map({'Ghost': 1, 'Goblin': 2, 'Ghoul': 3})
```

```python
df.groupby('type').type.count().plot.bar()
plt.xlabel('Class')
plt.ylabel('Frequency')
```

```python
pd.options.display.max_columns = 24
df.describe(include='all')
```

```python
print(df.isnull().mean())
df[pd.isnull(df).any(axis=1)]
```

```python
df = df.interpolate()
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

```python
dataset_with_dummies = pd.get_dummies(df,prefix_sep='--') #One Hot Encoding
print(dataset_with_dummies.head())
dataset_with_dummies.to_csv('d1.csv')
```
